```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as web
```

```python
def get_High(ticker, start, end):
    
    start = start
    end = end
    info = web.DataReader(ticker, data_source='yahoo', start=start, end=end)['High']
    return pd.DataFrame(info)
```

```python
AAPL = get_High('AAPL', '1/2/2020', '31/05/2021')
MSFT = get_High('MSFT', '1/2/2020', '31/05/2021')
```

```python
AAPL[['High']].plot(figsize=(12,6))
plt.title('APPLE high price analysis')
plt.ylabel('Price (USD)')
MSFT[['High']].plot(figsize=(12,6))
plt.title('Microsoft high price analysis')
plt.ylabel('Price (USD)')
plt.show();
```
