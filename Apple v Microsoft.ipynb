```python
import numpy as np
import pandas as pd
import io
import requests
import matplotlib.pyplot as plt
```

```python
url="https://query1.finance.yahoo.com/v7/finance/download/MSFT?period1=1590599403&period2=1622135403&interval=1d&events=history&includeAdjustedClose=true"
s=requests.get(url).content
print(s)
```

```python
microsoft = pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
microsoft.info()
```

```python
microsoft.head()
```

```python
url="https://query1.finance.yahoo.com/v7/finance/download/AAPL?period1=1590604174&period2=1622140174&interval=1d&events=history&includeAdjustedClose=true"
a=requests.get(url).content
print(a)
```

```python
apple = pd.read_csv(io.StringIO(a.decode('utf-8')))
```

```python
apple.info()
```

```python
apple.head()
```

```python
plt.plot(microsoft.Date, microsoft.High, 'c', label ='High price')
plt.plot(microsoft.Date, microsoft.Low, 'r', label ='Low price')
plt.title("Microsoft stock price dynamics")
plt.xlabel('Date')
plt.ylabel('Stock price')
plt.legend()
plt.show()
plt.savefig('microsoft.png' )

plt.plot(apple.Date, apple.High, 'c--', label ='High price')
plt.plot(apple.Date, apple.Low, 'r--', label ='Low price')
plt.title("Apple stock price dynamics")
plt.xlabel('Date')
plt.ylabel('Stock price')
plt.legend()
plt.show()
plt.savefig('apple.png')

```
