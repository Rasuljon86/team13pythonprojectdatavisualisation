```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pd.read_csv('A_stock_trends_17_21.csv')
```

```python
apple=pd.read_csv('A_stock_trends_17_21.csv')
type(apple)
```

```python
apple['close'].mean()
```

```python
apple['close'].median()
```

```python
apple['close'].plot(kind='box', vert=False, figsize=(14,6))
```

```python
apple['close'].plot(kind='density', figsize=(14,6))
```

```python
ax = apple['close'].plot(kind='density', figsize=(14,6)) 
ax.axvline(apple['close'].mean(), color='red')
ax.axvline(apple['close'].median(), color='green')
```

```python
apple['average_price'] = (apple['high'] + apple['low']) /2

apple['average_price'].head()
```

```python
apple['revenue_blns'] = apple['close'] * apple['volume']/10**6

apple['revenue_blns'].head()
```

```python
(apple['revenue_blns'] != apple['revenue_blns']).sum()
```

```python
apple.describe()
```

```python
ax = apple['average_price'].plot(kind='hist', figsize=(14,6))
ax.set_ylabel('Number of Sales')
ax.set_xlabel('dollars')
```

```python
corr = apple.corr()

corr
```

```python
apple.plot(kind='scatter', x='average_price', y='revenue_blns', figsize=(6,6))
```

```python
boxplot_cols = ['date', 'open', 'high', 'low', 'close', 'volume', 'revenue_blns']

apple[boxplot_cols].plot(kind='box', subplots=True, layout=(2,3), figsize=(14,8))
```

```python
apple['average_price'].plot(kind='hist', figsize=(14,6))
```

```python
plt.figure(figsize=(14,6), dpi=1000)

plt.plot(apple.date, apple.high, 'r', label ='High price')
plt.plot(apple.date, apple.low, 'y', label ='Low price')
plt.plot(apple.date, apple.average_price, 'g', label ='Average price')
plt.title("Apple stock price dynamics during January 2017-June 2021")
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('Date')
plt.ylabel('Stock price')
plt.legend()
plt.show()
plt.savefig('apple_stock.png' )


plt.show()
```

```python
plt.figure(figsize=(14,6), dpi=1000)

plt.plot(apple.average_price, apple.revenue_blns, 'rx', label ='High price')
plt.title("Stock price to revenue during for 2017-2021 yy")
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('Average stock price')
plt.ylabel('Revenue')
plt.legend()
plt.show()
plt.savefig('apple_stock_revenue.png' )


plt.show()
```
